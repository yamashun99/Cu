```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dos = pd.read_csv(
    "cu.dos", sep="\s+", names=["energy", "dos", "idos"], skiprows=1, header=None
)
```

```python
cu_d = pd.read_csv(
    "cu.pdos_atm#1(Cu)_wfc#3(d)",
    sep="\s+",
    names=["energy", "ldos", "pdos1", "pdos2", "pdos3", "pdos4", "pdos5"],
    skiprows=1,
    header=None,
)
```

```python
ef = 13.3289
fig, ax = plt.subplots()
ax.plot(dos["energy"] - ef, dos["dos"], label="DOS")
ax.plot(cu_d["energy"] - ef, cu_d["ldos"], label="Co LDOS")
```
